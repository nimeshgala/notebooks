```
def get_momentum_studies2(df, periods = 13):
  open = df['Open']
  low = df['Low']
  high= df['High']
  close= df['Adj Close']
  volume= df['Volume']
  
  # Momentum studies
  # https://mrjbq7.github.io/ta-lib/func_groups/momentum_indicators.html
  
  df['macd'], df['macd_sign'], df['machd_hist'] = talib.MACD(close, fastperiod=12, slowperiod=26, signalperiod=9)
  df['aroon_down'], df['aroon_up'] = talib.AROON(high, low, timeperiod=periods)
  df['ema_13'] = talib.EMA(close, timeperiod=13)
  df['ema_21'] = talib.EMA(close, timeperiod=21)
  df['ema_144'] = talib.EMA(close, timeperiod=144)
  df['ema_199'] = talib.EMA(close, timeperiod=199)
  
  df["minus_di"] = talib.MINUS_DI(high, low, close, timeperiod=periods)
  df["plus_di"] = talib.PLUS_DI(high, low, close, timeperiod=periods)
  df["adx"] = talib.ADX(high, low, close, timeperiod=periods)
  df["daily_rsi"] = talib.RSI(close, timeperiod=periods)
  df['weekly_rsi'] = talib.RSI(close, timeperiod=periods*5)
  df['monthly_rsi'] = talib.RSI(close, timeperiod=periods*21)
  
  df['vwap'] = np.cumsum(volume*(high+low)/2) / np.cumsum(volume)
```

```
def get_rsi_score(daily_rsi, weekly_rsi):
  if(daily_rsi > 55 and daily_rsi <= 60) or (daily_rsi < 45 and daily_rsi >= 40):
    target_size = 1;
  elif(daily_rsi > 60 and daily_rsi <= 65) or (daily_rsi < 40 and daily_rsi >= 35):
    target_size = 2;
  elif(daily_rsi > 65 and daily_rsi <= 70) or (daily_rsi < 35 and daily_rsi >= 30):
    target_size = 3;
  elif(daily_rsi > 70) or (daily_rsi < 30):
    target_size = 4
  else:
    target_size = 0;
  
  if(daily_rsi < 45):
    target_size = target_size*-1

  if((daily_rsi > weekly_rsi) and (weekly_rsi > 50)):
    target_size = target_size + 2
  elif((daily_rsi < weekly_rsi) and (weekly_rsi < 50)):
    target_size = target_size - 2
    
  return target_size
```

```
def get_ema_score(close, ema_13, ema_21, ema_144, ema_199):
  if (close > ema_13) and (ema_13 > ema_21) and (ema_21 > ema_144) and (ema_144 > ema_199):
    return 4
  elif (close > ema_13) and (ema_13 > ema_21):
    return 2
  
  if (close < ema_13) and (ema_13 < ema_21) and (ema_21 < ema_144) and (ema_144 < ema_199):
    return -4
  if (close < ema_13) and (ema_13 < ema_21):
    return -2

  return 0;
```

```
def calculate_score(row):
  close = row['Adj Close']
  max_pain = None 
  vwap = row['vwap']
  oi_pcr_diff = None #row['vwap']
  daily_rsi = row['daily_rsi']
  weekly_rsi = row['weekly_rsi']
  adx = row['adx']
  adx_diff = row['adx_diff']
  minus_di = row['minus_di']
  plus_di = row['plus_di']
  aroon_up = row['aroon_up']
  aroon_down = row['aroon_down']
  ema_13 = row['minus_di']
  ema_21 = row['ema_21']
  ema_144 = row['ema_144']
  ema_199 = row['ema_199']

  score = 0

  score = get_ema_score(close, ema_13, ema_21, ema_144, ema_199)

  if(max_pain is not None and max_pain is not np.NaN):
    if close >  max_pain:  
      score = score +1 
    elif close < max_pain:
      score = score -1;
  if(oi_pcr_diff is not None and oi_pcr_diff is not np.NaN):
    if oi_pcr_diff > 0 :
      score = score +1
    elif oi_pcr_diff < 0:
      score = score -1
  
  if(close >  vwap):  
    score = score +1 
  elif close < vwap: 
    score = score -1;

  score = score + get_rsi_score(daily_rsi, weekly_rsi)

  adx_score = 0
  if(adx > 25):
    adx_score = 2
  elif adx > 20:
    adx_score = 1

  if adx_diff > 0:
    score = score + adx_score
  elif adx_diff < 0:
    score = score - adx_score

  if(plus_di > minus_di):
    score = score +1
  elif(plus_di < minus_di):
    score = score -1

  if(aroon_up > 50):
    score = score + 1
  elif aroon_down > 50:
    score = score - 1

  return score;
```
